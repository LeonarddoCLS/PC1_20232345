**Fourth task**



```python
#Creamos una lista con la raíz cuadrada de las emisiones de  CO2
carbon_root_of_metric_tonnes_CO2=carbon_copy.metric_tonnes_of_CO2.apply(lambda x: x**0.5)
carbon_root_of_metric_tonnes_CO2
```

```python
#agregamos una nueva columna carbon_root_of_metric_tonnes_CO2
carbon_copy['root_of_metric_tonnes_of_CO2']=carbon_root_of_metric_tonnes_CO2
carbon_copy
```


**Fifth Task**


```python
#Calculamos el promedio de las medidas de toneladas de CO2
carbon_copy.metric_tonnes_of_CO2.mean()
```

```python
#Creamos una nueva columna en donde se reste las medidas de toneladas de CO2 con el promedio
carbon_copy['metric_tonnes_of_CO2_minus_average']=carbon_copy.metric_tonnes_of_CO2-carbon_copy.metric_tonnes_of_CO2.mean()
carbon_copy
```
